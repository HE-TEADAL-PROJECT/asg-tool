This notebook uses the vectorestore created in the load_vs_sample_notebook.
It uses the function generate from acg.agents.build_connector to build a connector for the
query we define.
And executes (from acg.executor) the connector to get the response.
Need to run the fastAPI server locally to get the request and forward it to Teadal's endpoints and print the response.

```python
from acg import config
from acg.agents import build_connector
from acg.executor import exec
from acg.util.logging import Logging
```

```python

# Used the persistant vectorstore - fdp-amts-gtfs

query = "get all stops"
```

```python

base_url = "http://localhost:8001/"
connector_file = "fdp-amts-gtfs-static-connector.yaml"
config_file = "../persistent_vectorstore_config.yaml"
```

```python

build_connector.generate(base_url,
                            config_file,
                            None,
                            connector_file,
                            query)
```

```python
import pprint
import pandas as pd
# Run the connector and return a pandas data frame
out = exec.run_from_spec_file(connector_file)
```

```python

df = pd.json_normalize(out, 'data')
filtered_df = df.query('stop_id == 100')
pprint.pprint(filtered_df)
```
