This notebook shows how to load an OpenAPI spec file to a persistent vectorestore
using the connectors flow1 code which is imported as a library. 
After running this code sample, vectorestore will be created under notebooks/var.

```python

from acg import config
from acg.retrieval import vectorstore_loader
```

```python
# Make sure the following paths are correct
config_file = "../persistent_vectorstore_config.yaml"
conf = config.import_config(config_file)

specs_files = [ "../../openapi-specs/fdp-amts-gtfs-static.yaml"]

# Load into persistant store to keep APIs
vectorstore_loader.add_specs(conf, specs_files)
```
